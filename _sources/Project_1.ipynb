## Part 2: Calculate Statistics with Pandas

Now let's calculate the three measures of central tendency:
- **Mean**: The average rating
- **Median**: The middle value when sorted
- **Mode**: The most frequent rating

```python
# Calculate statistics using pandas
mean_pandas = selected_column.mean()
median_pandas = selected_column.median()
mode_result = selected_column.mode()

# Handle mode (might have multiple modes)
if len(mode_result) > 0:
    mode_pandas = mode_result[0]  # take first mode
else:
    mode_pandas = None

# Display results
print("Statistics from Pandas:")
print(f"Mean:   {mean_pandas:.4f}")
print(f"Median: {median_pandas:.4f}")
print(f"Mode:   {mode_pandas}")

# Let's see what these numbers tell us
print(f"\nInterpretation:")
print(f"The average rating of {mean_pandas:.2f} suggests...")
if mean_pandas > 4:
    print("  - Investors have a positive view of Mutual Funds")
else:
    print("  - Investors are neutral/cautious about Mutual Funds")
```

```python
# Check the distribution of ratings
rating_counts = selected_column.value_counts().sort_index()
print("Distribution of ratings:")
for rating, count in rating_counts.items():
    percentage = (count / len(selected_column)) * 100
    print(f"Rating {rating}: {count} responses ({percentage:.1f}%)")
```

## Part 3: Manual Calculations 

Now let's calculate the same statistics WITHOUT using pandas or any 
statistics library. This helps us understand what's happening behind 
the scenes.

```python
# Read data manually
import csv

data = []
with open("Finance_Trends.csv", "r") as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        data.append(int(row["Mutual_Funds"]))

print(f"Read {len(data)} values")
```

```python
# Mean = sum / count
mean_manual = sum(data) / len(data)
print(f"Mean: {mean_manual:.4f}")
```

```python
# Median = middle value
sorted_data = sorted(data)
n = len(sorted_data)

if n % 2 == 1:
    median_manual = sorted_data[n // 2]
else:
    median_manual = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2

print(f"Median: {median_manual:.4f}")
```

```python
# Mode = most frequent value (MUST USE DICTIONARY!)
counts = {}
for value in data:
    if value in counts:
        counts[value] += 1
    else:
        counts[value] = 1

# Find most common
mode_manual = max(counts, key=counts.get)
print(f"Mode: {mode_manual}")
```

## Part 4: Data Visualization

Create visualizations using only the standard library - no matplotlib!
We'll use ASCII characters to create visual representations.

```python
# Text-based visualization (STANDARD LIBRARY ONLY) for the column 'Mutual_Funds'
# - Clear title
# - X-axis label with scale
# - Renders a horizontal ASCII bar chart + sparkline

import csv
from collections import Counter
from math import ceil, isclose

CSV_FILE = "Finance_Trends.csv"
COL = "Mutual_Funds"  # chosen rating column (1..7 integers)


def as_1to7(s):
    s = (s or "").strip()
    if s.isdigit():
        n = int(s)
        return n if 1 <= n <= 7 else None
    try:
        x = float(s)
        if 1 <= x <= 7 and isclose(x, round(x), abs_tol=1e-9):
            return int(round(x))
    except:
        pass
    return None


# read ratings
vals = []
with open(CSV_FILE, newline="", encoding="utf-8") as f:
    for row in csv.DictReader(f):
        n = as_1to7(row.get(COL, ""))
        if n is not None:
            vals.append(n)
assert vals, f"No 1..7 ratings found in '{COL}'."

# distribution
cnt = Counter(vals)
dist = [cnt.get(k, 0) for k in range(1, 8)]
N, peak = len(vals), max(dist)

# chart params
WIDTH = 44  # fits narrow screens; tweak 36â€“60 if needed
unit = max(1, ceil(peak / WIDTH))  # each block â‰ˆ unit responses
scale = lambda c: max(1, c // unit) if c > 0 else 0

title = f"Rating Distribution â€” column: {COL}  (n={N})"
x_label = f"Responses (each â–ˆ â‰ˆ {unit})"

print("\n" + title)
print("=" * len(title))

# simple ruler (top)
tick_every = 8
ruler = "".join(("|" if i % tick_every == 0 else "â”€") for i in range(WIDTH))
ticks = [unit * i * tick_every for i in range(WIDTH // tick_every + 1)]
print("     " + ruler)
print("     " + "".join(str(t).ljust(tick_every) for t in ticks))

# bars
for k, c in enumerate(dist, 1):
    bar = "â–ˆ" * scale(c)
    pct = c / N * 100
    print(f"{k:>5} â”‚ {bar:<{WIDTH}}  {c:>5}  ({pct:>5.1f}%)")

print("     " + ruler)
print("     " + x_label)

# sparkline
levels = "â–â–‚â–ƒâ–„â–…â–†â–‡â–ˆ"


def level(x, x_max):
    if x_max == 0:
        return levels[0]
    i = int(round((len(levels) - 1) * x / x_max))
    return levels[i]


spark = "".join(level(c, peak) for c in dist)
print("\nSparkline (scores 1â†’7):")
print(" ".join(str(i) for i in range(1, 8)))
print(spark)

# brief interpretation
mode_score = 1 + dist.index(max(dist))
share = max(dist) / N * 100
print(
    f"\nInterpretation: score {mode_score} is most frequent (~{share:.1f}% of responses)."
)
```
